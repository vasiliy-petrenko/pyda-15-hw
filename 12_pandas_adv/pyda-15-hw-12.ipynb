# Задание 1

```python
csv_data = r'c:\Users\petrenkovn\Dropbox\DS-22\Data_Files\ml-latest-small\ratings.csv'

dataset = pd.read_csv( csv_data )

def classify_movie( rating ):
    """
    Функция классифицирует фильмы по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг
    """
    if rating <=2:
        rank = 'низкий рейтинг'
    elif rating <=4:
        rank = 'средний рейтинг'
    else:
        rank = 'высокий рейтинг'
    
    return rank

# результат записывается в столбец class
dataset['class'] = dataset['rating'].apply(classify_movie)

# проверка результатов работы функции - агрегация до class и rating
dataset[['class','rating','movieId']].groupby(['class','rating']).count()
```

```python
import re
```

```python
csv_data = r'c:\Users\petrenkovn\Dropbox\DS-22\Data_Files\keywords.csv'

dataset = pd.read_csv( csv_data )

dataset
```

```python
geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}

# подготовка обратного словаря Город-Регион и списка всех городов, разделенного |
geo_dict = {}
cities_regex = ''
for item in geo_data.items():
    for city in item[1]:
        geo_dict[city] = item[0]
        if len(cities_regex) > 0:
            cities_regex += '|' + city
        else:
            cities_regex = city
# print(geo_dict, cities)
            
def region_by_city( text ):
    """
    Функция выполняет задачу геоклассификации - в передаваемой текстовой строке выполняет поиск Города из ранее указанного словаря, и в случае если он найден - возвращает Регион
    """
    # используется поиск с помощью регулярных выражений
    result = re.findall( cities_regex, text )
    if len(result) > 0:
        city = result[0]
    else:
        city = None
    
    region = geo_dict.get(city, 'undefined')   
    return region

# применение функции к ряду датафрейма, запись в отдельный столбец
dataset['region'] = dataset['keyword'].apply( region_by_city )

# проверка корректности поиска
dataset[ dataset['region'] != 'undefined' ].head()
```
