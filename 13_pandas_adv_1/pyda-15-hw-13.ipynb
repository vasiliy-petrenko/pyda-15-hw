## Задание 1
Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

- если источник traffic_source равен yandex или google, то в source_type ставится organic
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений

```python
log = pd.read_csv(r'c:\Users\petrenkovn\Dropbox\DS-22\Data_Files\visit_log.02.csv', sep=';')

# применение loc для фильтрации данных датафрейма и определения значений в новом столбце
log.loc[log['traffic_source'].str.contains('yandex|google'), 'source_type'] = 'organic'
log.loc[log['traffic_source'].str.contains('paid') & log['region'].str.contains('Russia'), 'source_type'] = 'ad'
log.loc[log['traffic_source'].str.contains('paid') & ~log['region'].str.contains('Russia'), 'source_type'] = 'other'
log.loc[pd.isnull(log['source_type']), 'source_type'] = log['traffic_source']

# проверка результатов работы - агрегация
log[['source_type','traffic_source','region','visit_id']].groupby(['source_type','traffic_source','region']).count()


# log.head()
```

## Задание 3
В датафрейме data создайте столбец lemmas, в котором вычислите леммы поисковых запросов из столбца keyword. Леммы должны иметь строковый тип.

```python
from pymystem3 import Mystem
```

```python
# search = 'проверки проверка проверку'
m = Mystem()
# lemmas = m.lemmatize(search)
# lemmas

def lemmas_from_query(query):
    lemmas = m.lemmatize(query)
    str_lemmas = ' '.join(lemmas)
    return str_lemmas.strip()

data = pd.DataFrame({
    'keyword': ['курс гривны к рублю', 'доллары в рубли', '100 долларов в рублях', 'курс рубля'],
    'shows': [125076, 114173, 97534, 53546],
})

data['lemmas'] = data['keyword'].apply(lemmas_from_query)

data.head()
```
