```python
csv_data = r'c:\Users\petrenkovn\Dropbox\DS-22\Data_Files\power.csv'

dataset = pd.read_csv( csv_data )

# список стран, упорядоченный по алфавиту, чтобы скопировать нужные значения
# dataset.sort_values(by='country')['country'].unique()

# Посчитать суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.
dataset[ ( dataset['country'].isin(['Latvia','Lithuania','Estonia']) ) & ( dataset['category'].isin([4,12,21]) ) & ( dataset['year'] >= 2005 ) & ( dataset['year'] <= 2010 ) &  ( dataset['quantity'] > 0 ) ]['quantity'].sum()
```
